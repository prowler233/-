```python
from torch import nn
import torch
# PyTorch 中 Conv2d 默认使用 Kaiming Uniform 初始化
# 也叫 He 初始化

conv = nn.LazyConv2d(64, kernel_size=3)
x = torch.randn(1, 3, 32, 32)
with torch.no_grad():
    _ = conv(x)

print("Conv2d 默认初始化 (Kaiming Uniform):")
print(f"权重范围: [{conv.weight.min():.4f}, {conv.weight.max():.4f}]")
print(f"权重均值: {conv.weight.mean():.6f}")
print(f"权重标准差: {conv.weight.std():.6f}")

# 偏置默认初始化为 0
print(f"偏置: {conv.bias}")  # tensor([0., 0., ...])
```
